# Problem Statement <a id='problem_statement'></a>

Though great progress has been made in general aviation safety over the years, flying general aviation aircraft is still one of the most dangerous modes of transportation. There are two ways to improve the safety of pilots and passengers in light airplanes: reducing the overall aircraft accident rate, and reducing the fraction of accidents that are fatal. 

For this independent project, I am doing exploratory analysis on the Aircraft Accident database provided by the Federal Aviation Administration (FAA) and the National Transportation Safety Board (NTSB), with the goal of determining the important factors that contribute to an aircraft's accident being fatal.  Next, I am creating a classifier to predict whether a given accident will be fatal based on characteristics of the aircraft, pilot, and general conditions of the flight before the accident. 

I hope to provide some actionable insight to policy makers, aircraft manufacturers, and pilots in general to improve general aviation safety.

# Data Sources<a id='data_sources'></a>
## Aircraft Accident Database
The primary source of information for this analysis is the FAA/NTSB Aircraft Accident Database. 
This database contains all of the data collected during the NTSB investigations into aircraft
accidents from 1982 to the present. The database also contains the conclusions that the NTSB
came to regarding the cause of the accident, timeline of events, etc. 
The Microsoft Access database can be downloaded from: https://app.ntsb.gov/avdata/

The accident database has 13 tables and over 200 total unique fields. The following document shows an overview of the tables and relationships:

[Aircraft Accident Database Overview](https://app.ntsb.gov/avdata/admsrel.pdf). 

Then a detailed architecture with the tables and fields in each table can be viewed here:

[Aircraft Accident Database Detailed Architecture](https://app.ntsb.gov/avdata/eadmspub.pdf)

## Related Aviation Databases
1. [Aviation Accident Dataset](https://www.ntsb.gov/_layouts/ntsb.aviation/index.aspx)
    1. Subset of data from the above Aircraft Accident Database


2. [Aircraft Registration Database](https://www.faa.gov/licenses_certificates/aircraft_certification/aircraft_registry/releasable_aircraft_download/)
    1. Set of several different csv files. Contains aircraft registration number, aircraft serial number, aircraft make & model, manufactured year, etc.


3. [FAA Pilot Database](https://www.faa.gov/licenses_certificates/airmen_certification/releasable_airmen_download/)
    1. Set of several related csv files. Contains pilot's license holder information: Rating (single-engine land, multi-engine land, etc.), commercial and instructor ratings, individual aircraft type ratings, etc.
    
  

## Flying Hours and Accident Statistics from GAMA
Flight Hours and Accident data is reported by the General Aviation Manufacturers Association (GAMA) in their yearly industry review publication, which is available online. I gathered the data from two publications, the most recent from [2018](https://gama.aero/wp-content/uploads/GAMA-2018-Annual-Report-1.pdf) and then the report from [2013](https://gama.aero/wp-content/uploads/GAMA-2013-Databook-Updated-LowRes.pdf), which included data going back to 1938. 

## General Aviation
General Aviation as defined by NTSB includes operations under Federal Aviation Regulations (FAR) Part 91, Part 91K, Part 125, Part 133, and Part 137 for the purpose of accident statistics. Only those parts have been included in this analysis as well.

```python
# Function to remove duplicates, set all values to uppercase, and to strip all entries of leading and trailing spaces
def strip_strings(df):
    df = df.loc[:,~df.columns.duplicated()]
    df = df.dropna(axis=1,how='all')
    for col in df.columns:
        if df[col].dtype=='O':
            df[col] = df[col].str.strip()
        if df[col].dtype=='O':
            df[col] = df[col].str.upper()
    return(df)
```

```python
# Apply that function to each of the tables from that database
df_events = strip_strings(df_events)
df_aircraft = strip_strings(df_aircraft)
df_engines = strip_strings(df_engines)
df_flight_crew = strip_strings(df_flight_crew)
df_flight_time = strip_strings(df_flight_time)
df_findings = strip_strings(df_findings)
df_occurrences = strip_strings(df_occurrences)
```

```python
# Remove any non-GA records from the events table. This will trickle down to all the other tables which always use
# df_events.
ga_far_parts = ['091','091K','125','133','137']
df_events = df_events.merge(df_aircraft[['ev_id','far_part']],how='left',left_on='ev_id',right_on='ev_id')
df_events = df_events[df_events['far_part'].isin(ga_far_parts)]
```

```python
# Set total flight hours values listed as above 95000 to NaN, as they are almost certainly mis-entries
df_flight_time['flight_hours'] = df_flight_time['flight_hours'].apply(lambda x: np.nan if x > 95000 else x)
```

## Aircraft Accident Trends Over Time <a id='aircraft_accident_trends_over_time'></a>

```python
def plot_accident_frequency_over_time(start_year = 1938, end_year = 2017):
    df = df_ga_flight_hours
    fig,ax = plt.subplots()
    fig.set_size_inches(8,6)
    x = df.loc[start_year:end_year].index
    y1 = df['num_accidents'].loc[start_year:end_year]
    y2 = df['num_fatal_accidents'].loc[start_year:end_year]
    ax.plot(x,y1)
    ax.plot(x,y2)
    ax.fill_between(x,y2,y1,alpha=0.5)
    ax.fill_between(x,0,y2,alpha=0.5)
    ax.legend(['Total Accidents','Fatal Accidents'])
    ax.set_title('General Aviation Accidents & Fatal Accidents Over Time\n{} - {}'.format(start_year,end_year))
    ax.set_ylabel('Frequency per Year')
    a=ax.set_xlabel('Year')
    return
plot_accident_frequency_over_time()
```

As can be seen, there is a significant spike in accident frequency in the mid to late 1940s, though from the next graph of accident rates, it'll be shown that the overall accident rate was still continuing down during that time. During that specific period there are simply far, far more general aviation flight hours. 

```python
def plot_accident_rate(start_year=1938, end_year=2017):
    df = df_ga_flight_hours
    fig, ax = plt.subplots()
    fig.set_size_inches(8,6)
    x = df.loc[start_year:end_year].index
    y1 = df.loc[start_year:end_year]['accident_rate']
    y2 = df.loc[start_year:end_year]['fatal_accident_rate']
    ax.plot(x,y1)
    ax.plot(x,y2)
    ax.fill_between(x,y2,y1,alpha=0.5)
    ax.fill_between(x,0,y2,alpha=0.5)
    ax.legend(['Total Accident Rate','Fatal Accident Rate'])
    ax.set_title('General Aviation Accident & Fatal Accident Rates Over Time\n{} - {}'.format(start_year,end_year))
    ax.set_ylabel('Accidents per 100,000 miles')
    a=ax.set_xlabel('Year')
    return
plot_accident_rate()
print('*General aviation flight hours from 1943 to 1945 are unavailable')
```

It can be seen that there was a drastic reduction in accident rates from 1938 to about 1980, after which point the slope of the reduction in accidents slowed. We will focus on 1982 to the present in the graph below.

```python
plot_accident_rate(1982,2017)
```

As can be seen, the accident rate still dropped by quite a significant amount over the period from 1982 to 2017, however it's not nearly as drastic as in the previous years. The fatal accident rate has decreased significantly as well.

```python
def plot_fatal_nonfatal_accident_ratio(start_year=1938, end_year=2017):
    df = df_ga_flight_hours
    fig, ax = plt.subplots()
    fig.set_size_inches(8,6)
    x = df.loc[start_year:end_year].index
    y1 = df.loc[start_year:end_year]['fatal_accident_rate']/df.loc[start_year:end_year]['accident_rate']
    ax.plot(x,y1)
    ax.set_ylim(bottom=0)
    ax.set_title('Ratio of Fatal Accidents to Total Accidents over time\n{} - {}'.format(start_year,end_year))
    ax.set_ylabel('Ratio')
    a=ax.set_xlabel('Year')
    return
plot_fatal_nonfatal_accident_ratio()
```

While the ratio of fatal to non-fatal accidents is quite noisy, you can see that the ratio has still increased over time. That's primarily due to the non-fatal accident rate decreasing significantly, and the fatal accident rate not keeping up.

```python
plot_fatal_nonfatal_accident_ratio(1982,2017)
```

Looking at more recent data, it's apparent just how noisy the data is. The ratio from 1982 to 2017 has bumps in it, but it's mostly stayed within the range of .175 to .225.

## Determining Factors Involved in Fatal Accidents vs Non-Fatal Accidents<a id='determining_factors_involved_in_fatal_accidents_vs_non-fatal_accidents'></a>

```python
df_events['fatal'] = df_events['ev_highest_injury'].apply(lambda x: 1 if x=='FATL' else 0)
avg_fatal_rate = df_events['fatal'].value_counts()[1] / (len(df_events['fatal']))
print('Average Rate of Fatal Accidents vs Total Accidents (1982-2017): ',avg_fatal_rate)
```

```python
# Function to merge dfs in the most common fashion
def merge_dfs(df2,field):
    df = df_events[['ev_id','ev_highest_injury']].merge(df2[['ev_id',field]],how='left',left_on='ev_id',right_on='ev_id')
    return(df)
```

```python
# Function to plot the most common type of comparison I'm doing in this analysis
def plot_ratio_by_field(df,field,name,sort=False):

    df_rate = df.groupby([field,'ev_highest_injury']).size().unstack(fill_value=0)
    df_rate['TOTL'] = df_rate.sum(axis=1)
    df_rate['FATL_Ratio'] = df_rate['FATL']/df_rate['TOTL']
    if sort:
        df_rate = df_rate.sort_values('FATL_Ratio',ascending=False)

    fig,ax = plt.subplots()
    fig.set_size_inches(8,6)
    df_rate['FATL_Ratio'].plot.bar(ax=ax)
    ax.set_title('Ratio of Fatal Accidents to Total Accidents by {}\n1982 - Present'.format(name),fontsize=16)
    ax.set_ylabel('Ratio',fontsize=14)
    ax.set_xlabel(name,fontsize=14)
    a=ax.set_xticklabels(df_rate.index,rotation=0)
    
    return(fig,ax)
```

```python
df = df_events[['ev_highest_injury','ev_nr_apt_loc']]
fig,ax = plot_ratio_by_field(df,'ev_nr_apt_loc','Whether Accident Occurred On or Off Airport',sort=True)
a=ax.set_xticklabels(['OFF AIRPORT','UNKNOWN','ON AIRPORT'])
```

As can be seen above, accidents occurring on airports are far, far less likely to be fatal than accidents occurring off airport.

```python
# Plot fatal accident ratio by month
month_rate = df_events[['ev_month','ev_highest_injury']].groupby(['ev_month','ev_highest_injury']).size().unstack(fill_value=0)
month_rate['TOTL'] = month_rate.sum(axis=1)
month_rate['FATL_rate'] = month_rate['FATL'] / month_rate['TOTL']

fig,ax = plt.subplots()
fig.set_size_inches(8,6)
month_rate['FATL_rate'].plot.bar()
ax.set_xticklabels(['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'],rotation=0)
ax.set_ylabel('Ratio')
a=ax.set_title('Ratio of Fatal Accidents to Total Accidents by Month\n1982 - Present')
```

There's a clear relationship between month of the year and the rate at which accidents are fatal. Though June and July have the lowest rate, the entire range from March to August have relatively lower values than the late fall and winter months. My intuition as a pilot is that this is most likely due to worse weather and fewer daylight hours during the winter, though we'll have to look at the data to see if it supports that.

```python
# Plot ratio of fatal accidents vs light conditions
light_cond_rate = (df_events[['light_cond','ev_highest_injury']].groupby(['light_cond','ev_highest_injury'])
                   .size()
                   .unstack(fill_value=0))
light_cond_rate['TOTL'] = light_cond_rate.sum(axis=1)
light_cond_rate['FATL_rate'] = light_cond_rate['FATL'] / light_cond_rate['TOTL']

category_names = {'NDRK':'NIGHT/DARK','NR':'NOT REPORTED','NITE':'NIGHT','DAYL':'DAYLIGHT','NBRT':'NIGHT/BRIGHT'}
light_cond_rate = light_cond_rate.rename(index=category_names).sort_values('FATL_rate',ascending=False)

fig, ax = plt.subplots()
fig.set_size_inches(8,6)
light_cond_rate['FATL_rate'].plot.bar()
ax.set_xlabel('')
ax.set_xticklabels(light_cond_rate.index,rotation=45,ha='right')
ax.set_ylabel('Ratio',fontsize=14)
a=ax.set_title('Ratio of Fatal Accidents to Total Accidents by Light Condition at time of Accident\n1982 - Present', fontsize=16)
```

Accidents are much more likely to be fatal when they occur at night, especially on dark nights.

```python
# Plot ratio of fatal accidents vs time of day.
df = df_events.loc[:,['ev_time','ev_highest_injury']]
df['ev_time'] = pd.cut(df['ev_time'],24)
time_rate = df.groupby(['ev_time','ev_highest_injury']).size().unstack(fill_value=0)
time_rate['TOTL'] = time_rate.sum(axis=1)
time_rate['FATL_rate'] = time_rate['FATL'] / time_rate['TOTL']

fig,ax = plt.subplots()
fig.set_size_inches(8,6)
x = np.arange(0,24,1)
y = time_rate['FATL_rate'].values
ax.bar(x,y,align='edge',width=1)
ax.set_xticks(x)
ax.set_ylabel('Ratio',fontsize=14)
ax.set_xlabel('Time (Military Standard)',fontsize=14)
a=ax.set_title('Ratio of Fatal Accidents to Total Accidents by Time of Day\n1982 - Present',fontsize=16)
```

Understandably, accidents which occur late at night and in the early hours of the morning are far more likely to be deadly. This chart matches the previous chart which compares the accident ratio to the light conditions.

# Learning Model
I'll start with a simple logistic regression, and choose an array of variables that I think are pertinent, and see how well the algorithm does.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.dummy import DummyClassifier
```

## Feature Selection and Cleaning

```python
events_features = ['ev_id','ev_nr_apt_loc','light_cond','far_part','fatal','wx_cond_basic']
aircraft_features = ['ev_id','acft_make','acft_model','homebuilt','type_fly','acft_category','fixed_retractable',
                    'num_eng']
df_accidents = df_events[events_features]
df_accidents = df_accidents.merge(df_aircraft[aircraft_features],how='left',left_on='ev_id',right_on='ev_id')
df_accidents['make_model'] = df_accidents['acft_make'] + ' ' + df_accidents['acft_model']
```

```python
# Convert selected columns to categorical data
fields_to_categories = ['ev_nr_apt_loc','light_cond','far_part','wx_cond_basic','acft_make',
                        'homebuilt','type_fly','acft_category','make_model','fixed_retractable','num_eng']
df_accidents[fields_to_categories] = df_accidents[fields_to_categories].astype('category')
```

```python
# Reduce total # of makes to look at. Cast all but top 100 makes as 'RARE'
top_make = df_accidents['acft_make'].value_counts()[:100]
df_accidents['acft_make'] = df_accidents['acft_make'].apply(lambda x: x if x in top_make else 'RARE')
```

```python
top_make_model = df_accidents['make_model'].value_counts()[:100]
df_accidents['make_model'] = df_accidents['make_model'].apply(lambda x: x if x in top_make_model else 'RARE')
# df_accidents['make_model'].value_counts()
```

```python
features = ['ev_nr_apt_loc','light_cond','far_part','wx_cond_basic','acft_make','homebuilt',
            'type_fly','acft_category','make_model','fixed_retractable','num_eng']
label = 'fatal'

X_train, X_test, y_train, y_test = train_test_split(df_accidents[features],df_accidents[label],random_state=0)
```

```python
X_train_dummies = pd.get_dummies(X_train)
X_test_dummies = pd.get_dummies(X_test)
```

## Baseline

```python
dummy_clf = DummyClassifier(strategy='most_frequent').fit(X_train,y_train)
# dummy_clf = DummyClassifier(strategy='stratified').fit(X_train,y_train)
dummy_predictions = dummy_clf.predict_proba(X_test)[:,1]
dummy_auc_score = roc_auc_score(y_test,dummy_predictions)
print('Dummy AUC Score: ',dummy_auc_score)
```

## Logistic Regression

```python
clf = Pipeline([
    ('scaler', MinMaxScaler()),
    ('clf', LogisticRegression(random_state=0,solver='liblinear'))  
])

parameters = {
    'clf__C':[.1,1,10,100]
}

gs_clf = GridSearchCV(clf,parameters,n_jobs=-1,cv=5)

gs_clf.fit(X_train_dummies.astype(float),y_train)
```

```python
predictions = gs_clf.predict_proba(X_test_dummies)
auc_score = roc_auc_score(y_test,predictions[:,1])

print('Train Score: ',gs_clf.score(X_train_dummies,y_train))
print('Test Score: ',gs_clf.score(X_test_dummies,y_test))
print('AUC Score: ',auc_score)
```

```python
df_coef = pd.DataFrame(index=X_train_dummies.columns,data=gs_clf.best_estimator_.named_steps['clf'].coef_[0])
df_coef.columns = ['Coefficient']
imp_coef = pd.concat([df_coef['Coefficient'].sort_values().head(10), df_coef['Coefficient'].sort_values().tail(10)])
fig,ax = plt.subplots()
imp_coef.plot(kind='barh')
fig.subplots_adjust(left=.3)
ax.set_xlabel('Model Coefficient')
a=ax.set_title('Largest Coefficients in Logistic Regression Model')
```

```python
# Plot the ROC Curve
fig, ax = plt.subplots()
fpr, tpr, _ = roc_curve(y_test,predictions[:,1])
ax.plot(fpr,tpr,label='ROC Curve')
ax.plot([0,1],[0,1],linestyle='--')
fig.set_size_inches(6,6)
ax.set_title('ROC Curve, Logistic Regression',fontsize=16)
ax.set_xlabel('False Positive Rate',fontsize=14)
ax.set_ylabel('True Positive Rate',fontsize=14)
a=ax.legend()
```
